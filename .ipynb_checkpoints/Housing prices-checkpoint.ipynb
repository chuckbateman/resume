```python
# Histogram of housing prices.  We see a normal distribution here centered around the median price of 225k
fig, ax = plt.subplots(figsize=(8,6))

ax.hist(x=df['Price'],
       bins=30, 
       edgecolor='white')

ax.set_ylabel('House Price')

plt.title('Housing prices')
;
```

```python
# I want to plot the median price by neighborhood.  This will require aggregating the data using the groupby function in pandas
df_neighbor = df.groupby(by='Neighborhood', as_index=False).median()
```

```python
# Median prices by neighborhood are very similar
sns.set_style('darkgrid')
sns.barplot(data=df_neighbor, x='Neighborhood', y='Price', estimator='median')
plt.title('Median House Prices by Neighborhood')
;
```

```python
# I want to feature a variable basedon the decade a house was built.  This simple function will help.
def decade(year):
    if year < 1960:
        return '1950s'
    elif year < 1970:
        return '1960s'
    elif year < 1980:
        return '1970s'
    elif year < 1990:
        return '1980s'
    elif year < 2000:
        return '1990s'
    elif year < 2010:
        return '2000s'
    elif year < 2020:
        return '2010s'
    else:
        return '2020s'
```

```python
df['decade built'] = df['YearBuilt'].apply(decade)

df = df.sort_values(by='decade built') 
```

```python
# Even after grouping by decade the price does not seem to fluxuate much from the decade a home was built in
sns.barplot(data=df, x='decade built', y='Price', estimator='median', ci=None)
plt.title('Median House Prices by Decade Built')
;
```

```python
# When counting by decade built we have a pretty even number for full decades but we only have 2020 and 2021 for the 2020s so far
sns.countplot(data=df, x='decade built')
plt.title('Count of Houses by Decade Built')
;
```

```python
# Using boxplots to compare house prices by number of rooms
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Bedrooms', y='Price')
plt.title('House Price Distribution by Number of Bedrooms')
;
```

```python
# Looking at the relationship between variables using matplotlib and the only real sign of correlation with price is square feet
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,6))

ax[0][0].scatter(x=df['SquareFeet'], y=df['Price'], alpha=.05)
ax[0][1].scatter(x=df['Bedrooms'], y=df['Price'])
ax[1][0].scatter(x=df['Bathrooms'], y=df['Price'])
ax[1][1].scatter(x=df['YearBuilt'], y=df['Price'], alpha=.05)

ax[0][0].set_title('Square Feet')
ax[0][1].set_title('Bedrooms')
ax[1][0].set_title('Bathrooms')
ax[1][1].set_title('Year Built')

ax[0][0].set_ylabel('Price')
ax[1][0].set_ylabel('Price')

plt.tight_layout()
```

```python
# Here I use a correlation matrix to confirm the correlation between price and square feet.
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap="Blues")
plt.title('Correlation heatmap')
```

```python
# Another simpler way to view relationships between variables is using pairplot
# This function is resource intensive and I often don't use it when dealing with large datasets
sns.pairplot(data=df, corner=True)
;
```

### Regression analysis

```python
# creating new dataframe with dummy variables for neighborhood and decade built
df_m = pd.get_dummies(df)

df_m.head()
```

```python
# I will include all variables on my first model and see how it performs.
features = ['SquareFeet', 'Bedrooms', 'Bathrooms', 'YearBuilt', 
       'Neighborhood_Rural', 'Neighborhood_Suburb', 'Neighborhood_Urban',
       'decade built_1950s', 'decade built_1960s', 'decade built_1970s',
       'decade built_1980s', 'decade built_1990s', 'decade built_2000s',
       'decade built_2010s', 'decade built_2020s']

X = sm.add_constant(df_m[features])
y = df_m['Price']

model = sm.OLS(y, X).fit()

model.summary()
```

```python
# The dummy variables (decade built and neighborhood) and year built are not statistically significant so I will drop them
features = ['SquareFeet', 'Bedrooms', 'Bathrooms']

X = sm.add_constant(df[features])
y = df['Price']

model = sm.OLS(y, X).fit()

model.summary()
```

#### Our R2 score is still pretty bad.  I will try a different algorithm and see how it performs

```python
# Data Splitting
features = ['SquareFeet', 'Bedrooms', 'Bathrooms']

X = sm.add_constant(df[features])
y = df["Price"]

X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
#standardization
std = StandardScaler()
X_tr = std.fit_transform(X.values)
X_te = std.transform(X_test.values)
```

```python
# Testing out different alphas and l1 ratios for elasticnet with gridsearch
n_alphas = 200
l1_ratios = np.linspace(.01, 1, 100)
alphas = 10 ** np.linspace(-3, 3, n_alphas)

enet_model = ElasticNetCV(alphas=alphas, l1_ratio=l1_ratios, cv=5)

enet_model = enet_model.fit(X_tr, y)

print(f"Tuned Alpha: {enet_model.alpha_}")
print(f"Tuned Lambda: {enet_model.l1_ratio_}")
print(f"Training Score: {enet_model.score(X_tr, y)}")
print(f"Test Score: {enet_model.score(X_te, y_test)}")
```

These results are no better than the basic OLS model and unfortunately neither model would be great for predicting home prices.  The biggest issue with is with the data which was synthetically created.  If I were to run this analysis with real life data, I would look for more variables such as the size of the property, median household income in the city where the house was built, city population and school ratings in the area.
